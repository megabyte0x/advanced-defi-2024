```python
import numpy as np
import matplotlib.pyplot as plt

### EMA - regular interval ###
def ema(p, u):
    a = 0.1
    return a*p + (1 - a)*u

### Plot ###
N = 100
# initial price
p0 = 10
# volatility
v = 2
delta_prices = np.random.normal(0, v, N)

# prices
p = p0
prices = []
for dp in delta_prices:
    prices.append(p)
    p += dp
    if p < 0:
        p = 0

# moving averages
emas = []
u = p0
for i in range(N):
    emas.append(u)
    p = prices[i]
    u = ema(p, u)

plt.plot(prices)
plt.plot(emas)
plt.legend(["price", "ema"])
plt.show()
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

### EMA - irregular interval ###
# half life
H = 4

# 1 - a_dt = 0.5**(dt/H) = e**(ln(0.5)/H * dt)
# 0.5**(dt/H) = (1 - a)**dt, a = 1 - e**(ln(0.5)/H)

# Find a with half life H
# 0.5**(dt/H) = (1 - a)**dt
# dt/H * ln(0.5) = dt * ln(1 - a)
# ln(0.5) / H = ln(1 - a)
# e**(ln(0.5) / H) = 1 - a
# a = 1 - e**(ln(0.5)/H)
def ema(p, u, dt):
    a = 1 - math.exp(math.log(0.5) / H * dt)
    return a*p + (1 - a)*u

### Plot ###
N = 100
# initial price
p0 = 10
# volatility
v = 2
delta_prices = np.random.normal(0, v, N)
# delta_prices = [-1 for i in range(N)]

# prices
p = p0
prices = []
for dp in delta_prices:
    prices.append(p)
    p += dp
    if p < 0:
        p = 0

# moving averages
# avgs[0] = regular interval
# avgs[1] = irregular interval
emas = [[], []]

# regular interval
u = p0    
for t in range(N):
    emas[0].append(u)
    p = prices[t]
    u = ema(p, u, 1)

# irregular interval
t = 0
ts = []
u = p0
while t < N:
    ts.append(t)
    emas[1].append(u)

    dt = np.random.randint(1, 2*H)
    t += dt
    
    if t < N:
        p = prices[t]
        u = ema(p, u, dt)

plt.plot(prices)
plt.plot(emas[0], marker = "+")
plt.plot(ts, emas[1], marker="o")
plt.legend(["price", "regular interval ema", "irregular interval ema"])
plt.show()
```
